```python
import pandas as pd

# Load the dataset
penguins = pd.read_csv('penguins.csv')
penguins.info()
```

```python
# Display the first 10 rows
penguins.head(10)
```

```python
# Display the datatype of each column
penguins.dtypes
```

```python
# Count the number of 'Species'
target_column = 'Species'
penguins[target_column].value_counts()
```

```python
import seaborn as sns

# Visualize the relationship between Culmen Length and Culmen Depth
row_count = penguins.shape[0]

_ = sns.pairplot(
    data = penguins[:row_count],
    vars = ['Culmen Length (mm)', 'Culmen Depth (mm)'],
    hue = 'Species',
    plot_kws = {'alpha': 0.4},
    height = 3,
    diag_kind = 'hist',
    diag_kws = {'bins': 30},
    )
```
