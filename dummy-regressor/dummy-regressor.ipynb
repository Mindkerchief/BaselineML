```python
import pandas as pd

canucks_df = pd.read_csv('canucks.csv')
canucks_df.info()
canucks_df.head()
```

```python
# Fill the missing values
canucks_df['Salary'] = canucks_df['Salary'].fillna(canucks_df['Salary'].mean(),)
canucks_df['Salary'].isna().sum()
```

```python
# Separate the features
x = canucks_df.loc[:,['No.', 'Age', 'Height', 'Weight', 'Experience']]
x.head()
```

```python
# Separate the target variable
y = canucks_df['Salary']
y.head()
```

```python
from sklearn.dummy import DummyRegressor

# Train the dummy regressor model
model = DummyRegressor(strategy='mean')
model.fit(x, y)
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy using score
accuracy_1 = model.score(x, y)
print(f'Accuracy using score: {accuracy_1:.2}')

# Calculate accuracy using accuracy_score
y_pred = model.predict(x)
accuracy_2 = accuracy_score(y, y_pred)
print(f'Accuracy using accuracy_score: {accuracy_2:.2}')
```
